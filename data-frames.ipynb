# Pandas has 2 data structures

## Series - one dimensional array with indices. It stores  a single column or single row of data from a data frame. The datatype of each value in a series is the same.

## Data Frame - spreadsheet like structure that can store values of different datatypes 

```python
import numpy as np
import pandas as pd
```

```python
Series_1 = pd.Series(np.random.rand(34))
```

```python
Series_1 # creates a Series of random values with indices from 0 to 34-1
```

```python
type(Series_1) # prints the type of the data structure
```

```python
Data_frame_1 = pd.DataFrame(np.random.rand(334, 5), index=np.arange(334))
# creates a dataframe with shape (334, 5) with indices from 0 to 334
```

```python
Data_frame_1
```

```python
Data_frame_1.head() # prints the first few rows of the dataframe
```

```python
Data_frame_1.tail() # prints the first few rows of the dataframe
```

```python
Data_frame_1.describe() # prints the statistical measures of the series with numerical values
```

```python
Data_frame_1.dtypes # prints the type of the columns in the dataframe
```

```python
Data_frame_1[0][0] = 'str' # changes the value at 0,0 to 'str'
```

```python
Data_frame_1.dtypes # the datatype of the first column changes to object since one of the floats has been converted to string
```

```python
Data_frame_1.index # prints the indices of the dataframe
```

```python
Data_frame_1.columns # prints the columns of the dataframe
```

```python
Data_frame_1[0][0] = 0.189005 # changes the value at 0,0 to 0.189005
```

```python
Data_frame_1.to_numpy() # converts the dataframe to a numpy array
```

```python
Data_frame_1.T # Transposes the dataframe i.e. the rows becomes the columns and the columns become the rows
# NOTE : This does not change Data_frame_1
```

```python
Data_frame_1.sort_index(axis=1, ascending=False) #sorts the dataframe in descending values  of column no.s
# ascending is true by default.
```

```python
Data_frame_1.sort_index(axis=0, ascending=False) #sorts the dataframe in descending values  of indices
# ascending is true by default.
```

```python
Data_frame_1[0] # prints the first column of the data frame
```

```python
type(Data_frame_1[0]) # prints the type of the column of the dataframe 
```
